# DeepDRiD Exploration
Quick EDA: class balance, image samples, resolution check. Run this after placing the dataset in `../data/`.

```python
import sys
sys.path.append('../src')

import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

train_df = pd.read_csv('../data/train.csv')
train_df['label'].value_counts().sort_index().plot(kind='bar')
plt.title('DR Grade Distribution (Train)')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for grade, ax in enumerate(axes):
    sample = train_df[train_df['label'] == grade].iloc[0]
    img = Image.open('../data/' + sample['img_path'])
    ax.imshow(img)
    ax.set_title(f'Grade {grade}')
    ax.axis('off')
plt.show()
```
